Análise de dados de rotina de trabalho com Uber

```python
import pandas as pd
```

```python
df = pd.read_csv('../Uber - Tabela.csv')
# display(df.head())
df.value_counts()
```

```python
# Clean the 'Lucro' column:
# 1. Remove 'R$' and strip any whitespace
# 2. Replace ',' with '.' for decimal conversion
# 3. Convert to numeric, coercing errors to NaN
df['Lucro'] = df['Lucro'].astype(str).str.replace('R\$', '', regex=True).str.strip().str.replace(',', '.', regex=True)
df['Lucro'] = pd.to_numeric(df['Lucro'], errors='coerce')

# Drop rows where 'Lucro' is NaN (which now includes '#DIV/0!' and any original non-numeric entries)
df_cleaned = df.dropna(subset=['Lucro'])

print('DataFrame after dropping rows with no content in Lucro column:')
display(df_cleaned.head())
```
